### Question 4

Consider the set

Write code to compute the number of subsets of $A$ with an even number of elements
(hint - use a `for` loop).


```python
# Question 4
n = 20
num_subsets_even_count = 0
for k in range(0, n + 1, 2):
    num_subsets_even_count += math.comb(n, k)
print(f"{num_subsets_even_count}")
```

### Question 5

Consider the set

$$A=\{0,...,19\}$$

Write code to compute the number of subsets of $A$ where the sum of elements is a multiple of 4.

Hint: First write a function that takes a set and return `True` iff it elements sum to a multiple of 4.

```python
# Question 5

def is_sum_multiple_of_4(subset):
    return sum(subset) % 4 == 0

n = 20
set_A = list(range(n))

num_subsets_sum_multiple_of_4 = 0

for k in range(n + 1):
    for subset_tuple in its.combinations(set_A, k):
        if is_sum_multiple_of_4(list(subset_tuple)):
            num_subsets_sum_multiple_of_4 += 1

print(f"{num_subsets_sum_multiple_of_4}")
```

### Question 9

Write code to compute the number of lattice paths from (0,0) to (12,13).

```python
# Question 9

import math

def count_paths(n_up, n_right):
    return math.comb(n_up + n_right, n_up)

n_right = 12
n_up = 13
num_lattice_paths = count_paths(n_up, n_right)
print(f"({n_right},{n_up}) is: {num_lattice_paths}")
```
